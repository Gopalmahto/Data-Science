```python
ipl['result'].value_counts()
```

```python
#finding the number of time toss win wrt each team
ipl['toss_winner'].value_counts()
```

```python
#extracting the record where a team won and bat first
bat_first=ipl[ipl['win_by_runs']!=0]
```

```python
bat_first.head()
```

```python
bat_first.shape
```

```python
#making histogram 
plt.figure(figsize=(5,4))
plt.hist(bat_first['win_by_runs'])
plt.title('Distributions of runs')
plt.xlabel('Runs')
plt.grid(True)
plt.show()
```

```python
bat_first.sort_values(['win_by_runs'],ascending=False)
```

```python
bat_first['winner'].value_counts()
```

```python
#making a bar plot for top 3 team with most wins after batting first
x=bat_first['winner'].value_counts()[0:3].keys()
y=bat_first['winner'].value_counts()[0:3]

plt.figure(figsize=(8,5))
plt.bar(list(x),list(y),color=['blue','yellow','red'],width=0.6,edgecolor='black')
plt.show()
```

```python
#making pie-chart for thr same
x=bat_first['winner'].value_counts().keys()
y=bat_first['winner'].value_counts()
plt.pie(list(y),labels=list(x),autopct='%0.1f%%')
plt.show()
```
