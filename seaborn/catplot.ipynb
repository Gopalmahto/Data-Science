```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0,data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

```python
sns.boxplot(data['tip'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=data)
```
