```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,2*y)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':',linewidth=2)
plt.plot(x,y2,color='y',linestyle='-',linewidth=3)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth=2)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linestyle='--',linewidth=3)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]

plt.scatter(x,a,marker='*',c='r',s=50)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[2,3,2,0,4,2,3,5,9]
plt.scatter(x,a,marker='*',c='r',s=200)
plt.scatter(x,b,c='g',s=100)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,color='r',linewidth=2)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.subplot(1,2,2)
plt.scatter(x,b,color='g',marker='*',linewidth=3)
plt.title('X vs Y')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
fruit=['Apple','orange','Mango','banana','pineapple']
quantity=[20,50,45,12,40]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','black','pink','green'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=1,autopct='%0.1f%%')
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```
